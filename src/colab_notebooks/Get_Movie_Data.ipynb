```python
from convokit import Corpus, download
corpus = Corpus(filename=download("movie-corpus"))
```

```python
import pandas as pd
import numpy as np
```

```python
convo = corpus.random_utterance()
```

```python
convo.text
```

```python
utter_ids = corpus.get_utterance_ids()
```

```python
utterances = []
for id in utter_ids:
  utterances.append(corpus.get_utterance(id).text)
```

```python
utterances
```

```python
with open("movie_data.txt", "w") as txt_file:
    txt_file.write("sentence|label\n")
    for line in utterances:
        words = line.split()
        if words:
          txt_file.write("".join(line) + "|" + line.split()[-1] + "\n")
        else:
          txt_file.write("".join(line) + "|\n")
```
